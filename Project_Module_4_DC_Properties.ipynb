```python
#Import lide and library needed

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
locationDCProperties = "DC_Properties.csv"
dfDCProperties = pd.read_csv(locationDCProperties)
```

```python
dfDCProperties.columns
```

```python
dfDCProperties.head()
```

```python
#rename columns

dfDCProperties.rename(columns={'BATHRM': 'Bathroom', 'Unnamed: 0': 'Line No'}, inplace=True)
```

```python
#validate changes
dfDCProperties.columns
```

```python
#make a scatterplot
sns.lmplot(x='ROOMS', y='PRICE', data=dfDCProperties)
```

```python
#add more details to scatterplot
sns.lmplot(x='ROOMS', y='PRICE', data=dfDCProperties, 
           fit_reg=False, #remove regression line
           hue='WARD')   #color by evolution stage
```

```python
dfDCProperties.dtypes
```

```python
#make a scatterplot
sns.lmplot(x='STORIES', y='PRICE', data=dfDCProperties)
```

```python
#make a scatterplot
sns.lmplot(x='PRICE', y='STORIES', data=dfDCProperties)
```

```python
#add more details to scatterplot
sns.lmplot(x='PRICE', y='SALE_NUM', data=dfDCProperties, 
           fit_reg=False, #remove regression line
           hue='WARD')   #color by evolution stage
```

```python
#change size of plot
plt.subplots(figsize=(13,7))

#violin plot shows the distribution of attack power for each type
sns.violinplot(x='PRICE', y='SALE_NUM', data=dfDCProperties)
```
